Employment Type

Remote Ratio

Partially Remote = 50 (hybrid),
Fully remote (more than 80%) = 100 (remote)


Company Size
Less than 50 employees = Small (s)
50 to 250 employees = Medium (M)
More than 250 employees = Large (L)

# Predicting Future Data Science Salaries.

```python
import sklearn as sk
```

```python
df
```

```python
target = df["salary_in_usd"]
x= df.drop(["salary_in_usd","work_year"],axis=1)
```

```python
x= pd.get_dummies(x)
x.head()
```

```python
#toscale the data
from sklearn.preprocessing import StandardScaler
```

```python
#scaling helps reduce the value of our data so it can fit into the model.
scaler = StandardScaler()
x = scaler.fit_transform(x)
x
```

```python
#to split our data into test and train we import the library below
from sklearn.model_selection import train_test_split
```

```python
#we split the data set here and into test and train where 30% is assigned to the test value
x_train, x_test, y_train, y_test= train_test_split(x,target,test_size=0.3,random_state=4)
```

```python
#model fitting time, importing the libraries neccessary
from sklearn.linear_model import LinearRegression
```

```python
model= LinearRegression()
```

```python
#to fit the model ontop of our training dataet
model.fit(x_train,y_train)
```

```python
#print the coefficient and intercept, where each coeffiecient represent each column
print(model.intercept_)
print(model.coef_)
```

```python
#to compute the r squared score, r squared close to 1 is good
print(model.score(x_train,y_train))
```

```python
#To check our predicions
predictions = model.predict(x_test)
```

```python
predictions
#predictions are not looking good
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
score=r2_score(predictions,y_test)
mae= mean_absolute_error(predictions,y_test)
mse= mean_squared_error(predictions,y_test)
rmse=np.sqrt(mse)
```

```python
print(score)
print(mae)
print(mse)
print(rmse)
#this model is poor, we'll try another model
```

# Trying out XGBoost

```python
from xgboost import XGBRegressor

XGB = XGBRegressor()

XGB.fit(x_train, y_train)

predictions = XGB.predict(x_test)
predictions
```
